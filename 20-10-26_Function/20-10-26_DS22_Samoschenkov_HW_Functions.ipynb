```python
# DS-22 Самощенков Сергей
# Домашнее задание к лекции "Функции"
# Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

# Общие требования к программе:
# код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
# в коде отсутствуют глобальные переменные (за исключением documents и directories);
# пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу (вводом команды "q").
```

```python
# Задание 1
#     Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру
#         Примеры работы:
#         1.
#         Введите команду:
#         p

#         Введите номер документа:
#         10006
#         Результат:
#         Владелец документа: Аристарх Павлов

#         2.
#         Введите команду:
#         p

#         Введите номер документа:
#         12345
#         Результат:
#         Документ не найден в базе

    # Пункт 2. Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится¶
    #     Примеры работы:
    #     1.
    #     Введите команду:
    #     s

    #     Введите номер документа:
    #     10006
    #     Результат:
    #     Документ хранится на полке: 2

    #     2.
    #     Введите команду:
    #     p

    #     Введите номер документа:
    #     12345
    #     Результат:
    #     Документ не найден в базе

    # Пункт 3. Пользователь по команде "l" может увидеть полную информацию по всем документам
        # Пример работы:
        # Введите команду:
        # l
        # Результат:

        # №: 2207 876234, тип: passport, владелец: Василий Гупкин, полка хранения: 1
        # №: 11-2, тип: invoice, владелец: Геннадий Покемонов, полка хранения: 1
        # №: 10006, тип: insurance, владелец: Аристарх Павлов, полка хранения: 2

    # Пункт 4. Пользователь по команде "ads" может добавить новую полку
    #     Примеры работы:
    #     1.
    #     Введите команду:
    #     ads

    #     Введите номер полки:
    #     10
    #     Результат:
    #     Полка добавлена. Текущий перечень полок: 1, 2, 3, 10.

    #     2.
    #     Введите команду:
    #     ads

    #     Введите номер полки:
    #     1
    #     Результат:
    #     Такая полка уже существует. Текущий перечень полок: 1, 2, 3.

    # Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
    #     Примеры работы:
    #     1.
    #     Введите команду:
    #     ds

    #     Введите номер полки:
    #     3
    #     Результат:
    #     Полка удалена. Текущий перечень полок: 1, 2.

    #     2.
    #     Введите команду:
    #     ds

    #     Введите номер полки:
    #     1
    #     Результат:
    #     На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: 1, 2, 3.

    #     3.
    #     Введите команду:
    #     ds

    #     Введите номер полки:
    #     4
    #     Результат:
    #     Такой полки не существует. Текущий перечень полок: 1, 2, 3.

# перечень всех документов
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
# перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def doc_main_property(doc_id, doc_prop):
    for item in documents:
        if item['number'] == str(doc_id):
            return item[doc_prop]
            break
    return 'Документ не найден в базе'

def doc_owner(doc_id):
    for item in documents:
        if item['number'] == str(doc_id):
            return item['name']
            break
    return 'Документ не найден в базе'

def doc_shelf(doc_id):
    for key, value in directories.items():
        for doc in value:
            if doc == str(doc_id):
                return key
                break
    return 'Документ не найден в базе'

def shelves():
    shelf_list=''
    for key in directories.keys():
        shelf_list = shelf_list + str(key) + ', '
    return 'Текущий перечень полок: ' + shelf_list[0:(len(shelf_list)-2)]

def shelf_ads(shelf_num):
    dir_len_old=len(directories)
    directories.setdefault(shelf_num,'')
    if dir_len_old==len(directories):
        print('Такая полка уже существует. ', shelves())
    else:
        print('Полка добавлена. ', shelves())

def shelf_exist(shelf_num):
    if shelf_num in directories:
        return True
    else:
        return False

def shelf_is_empty(shelf_num):
    if  len(directories[shelf_num])==0:
        return True
    else:
        return False

def shelf_delete(shelf_num):
    del(directories[shelf_num])

command=''
while command !='q':
    command = input('Введите команду:')
    
    if command == 'p':
        doc_num = input('Введите номер документа:')
        print('Результат:')
        print(f'Владелец документа: {doc_main_property(doc_num,"name")}')

    if command == 's':
        doc_num = input('Введите номер документа:')
        print('Результат:')
        print(f'Документ хранится на полке: {doc_shelf(doc_num)}')

    if command == 'l':
        print('Результат:')
        for item in documents:
            print(f'№: {item["number"]}, тип: {item["type"]}, владелец: {item["name"]}, полка хранения: {doc_shelf(item["number"])}')

    if command == 'ads':
        shelf_num = input('Введите номер полки:')
        print('Результат:')
        shelf_ads(shelf_num)

    if command == 'ds':
        shelf_num = input('Введите номер полки:')
        print('Результат:')

        if shelf_exist(shelf_num):
        
            if shelf_is_empty(shelf_num):
                shelf_delete(shelf_num)
                print('Полка удалена. ', shelves())
            else:
                print('На полке есть документа, удалите их перед удалением полки. ', shelves())
        
        else:
            print('Такой полки не существует. ', shelves())

```
